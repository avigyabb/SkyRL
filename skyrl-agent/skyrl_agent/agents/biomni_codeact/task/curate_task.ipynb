```python
for data_name in ['opentargets', 'pharmaprojects', 'gwas_catalog']:
    task = gwas_causal_gene(path = '/dfs/project/bioagentos/biomni_data/benchmark/', dataset = data_name, num_samples = 100000)
    print(len(task))
    import pandas as pd
    train_indices, val_indices = task.split(ratio = 0.8, seed = 42)

    train_rows = []
    for idx in train_indices:
        res = task.get_example(idx)
        train_rows.append(
            {
                "instance_id": idx,
                "prompt": res['prompt'],
                "task_name": f"gwas_causal_gene_{data_name}",
            }
        )

    df_train = pd.DataFrame(train_rows)

    val_rows = []
    for idx in val_indices:
        res = task.get_example(idx)
        val_rows.append(
            {
                "instance_id": idx,
                "prompt": res['prompt'],
                "task_name": f"gwas_causal_gene_{data_name}",
            }
        )

    df_val = pd.DataFrame(val_rows)

    df_train.to_parquet(f'/dfs/project/bioagentos/rl_data/gwas_causal_gene_{data_name}/train.parquet')
    df_val.to_parquet(f'/dfs/project/bioagentos/rl_data/gwas_causal_gene_{data_name}/val.parquet')
```
